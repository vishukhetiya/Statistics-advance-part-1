# 1. What is a random variable in probability theory?
#### A random variable is a variable that takes on numerical values determined by the outcome of a random phenomenon.

# 2. What are the types of random variables?
#### Random variables can be classified into two types: discrete and continuous.

# 3. What is the difference between discrete and continuous distributions?
#### Discrete distributions deal with countable outcomes, while continuous distributions deal with uncountable outcomes over an interval.

# 4. What are probability distribution functions (PDF)?
#### A probability distribution function (PDF) gives the probabilities of occurrence of different possible outcomes.

# 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
#### CDF represents the probability that a random variable takes on a value less than or equal to a specific value, while PDF represents the density of probabilities at each value.

# 6. What is a discrete uniform distribution?
#### A discrete uniform distribution is one in which all outcomes are equally likely over a finite set.

# 7. What are the key properties of a Bernoulli distribution?
#### A Bernoulli distribution models a single trial with two possible outcomes: success (1) or failure (0), with a single parameter p.

# 8. What is the binomial distribution, and how is it used in probability?
#### The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials.

# 9. What is the Poisson distribution and where is it applied?
#### The Poisson distribution models the number of events occurring in a fixed interval of time or space, often used for rare events.

# 10. What is a continuous uniform distribution?
#### A continuous uniform distribution has constant probability over a given range [a, b].

# 11. What are the characteristics of a normal distribution?
#### The normal distribution is symmetric, bell-shaped, and characterized by its mean and standard deviation.

# 12. What is the standard normal distribution, and why is it important?
#### The standard normal distribution is a normal distribution with a mean of 0 and standard deviation of 1. It's useful for comparing different data sets.

# 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
#### The CLT states that the distribution of the sample mean approximates a normal distribution as the sample size becomes large.

# 14. How does the Central Limit Theorem relate to the normal distribution?
#### It explains why many distributions tend to be normal, especially when aggregating many independent effects.

# 15. What is the application of Z statistics in hypothesis testing?
#### Z statistics are used to determine how far a data point is from the mean in standard deviations, useful in hypothesis testing.

# 16. How do you calculate a Z-score, and what does it represent?
#### Z = (X - μ) / σ. It shows how many standard deviations a data point is from the population mean.

# 17. What are point estimates and interval estimates in statistics?
#### Point estimates give a single value estimate of a parameter; interval estimates provide a range of values (confidence intervals).

# 18. What is the significance of confidence intervals in statistical analysis?
#### They help quantify the uncertainty in a point estimate, giving a likely range for the true population parameter.

# 19. What is the relationship between a Z-score and a confidence interval?
#### A Z-score is used to calculate confidence intervals when population variance is known.

# 20. How are Z-scores used to compare different distributions?
#### They help standardize and compare values from different normal distributions.

# 21. What are the assumptions for applying the Central Limit Theorem?
#### The CLT assumes independent samples, a sufficiently large sample size, and identically distributed variables.

# 22. What is the concept of expected value in a probability distribution?
#### Expected value is the weighted average of all possible values that a random variable can take.

# 23. How does a probability distribution relate to the expected outcome of a random variable?
#### It represents the long-run average result of the random variable over many trials.
